G

```python
# extracted
sub_df = df.loc[550:990, 'Gender':'Coughing of Blood']

# saved
sub_df.to_csv('sub_dataset.csv', index=False)

# statistical analysis
sub_df.describe()
```

H

```python
# column index extractor
fn = lambda name: next(filter(lambda i: df.columns[i] == name, range(len(df.columns))))

# extracted
sub_df_last = df.iloc[-20:, fn('Shortness of Breath'):fn('Frequent Cold')]

# saved
sub_df_last.to_csv('sub_last_dataset.csv', index=False)

# statistical analysis
sub_df_last.describe()
```
